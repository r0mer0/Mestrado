```python
import numpy as np
import cv2
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
from google.colab.patches import cv2_imshow
```

```python
# Read the image

image = cv2.imread('/content/jaguar.jpg') #--imread() helps in loading an image into jupyter including its pixel values
cv2_imshow(image)
```

```python
mat_y = np.array([ [  -1, -2, -1], 
                   [ 0, 0, 0], 
                   [ 1, 2, 1]])
mat_x = np.array([ [ 0, -1, 0], 
                   [ -1, 5, -1], 
                   [ 0, -1, 0]])
  
mat_z = np.array([ [ -1, -1, -1], 
                   [ -1, 8, -1], 
                   [ -1, -1, -1]])
mat_w = np.array([ [ 1, 2, 1], 
                   [ 2, 4, 2], 
                   [ 1, 2, 1]])*1/16


filtered_image = cv2.filter2D(image, -1, mat_y)
cv2_imshow(filtered_image)
filtered_image = cv2.filter2D(image, -1, mat_z)
cv2_imshow(filtered_image)
filtered_image = cv2.filter2D(image, -1, mat_x)
cv2_imshow(filtered_image)
filtered_image = cv2.filter2D(image, -1, mat_w)
cv2_imshow(filtered_image)
```
